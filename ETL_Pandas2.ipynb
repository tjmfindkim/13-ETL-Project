```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
# Store filepath in a variable
character_file = "Movies/marvel-wikia-data.csv"
```

```python
# Read first Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
character_df = pd.read_csv(character_file, encoding="ISO-8859-1")
character_df.head()
```

```python
# split the string on 'name' column into 'CHARACTER' and 'ALIAS' columns
character_df[['CHARACTER','ALIAS']] = character_df.name.str.split("\s+\(", expand=True, n=1)
character_df.head()
```

```python
columns = ['ID', 'ALIGN', 'EYE', 'HAIR', 'ALIVE', 'CHARACTER']
marvel_character_df = character_df[columns]
marvel_character_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
marvel_character_df.to_csv("Output/marvel_character_df.csv", index=False, header=True)
```

```python
# Store filepath in a variable
movies_metadata_file = "Movies/movies_metadata.csv"
```

```python
# Read second Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
movies_metadata_df = pd.read_csv(movies_metadata_file, low_memory=False, encoding="ISO-8859-1")
movies_metadata_df.head()
```

```python
movie_columns = ['id', 'original_title', 'revenue', 'title', 'vote_average', 'vote_count']
movie_metadata_df = movies_metadata_df[movie_columns]
movie_metadata_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
movie_metadata_df.to_csv('output/movie_metadata.csv', index = False, header=True)
```

```python
# Dependencies
import ast
def transform_cast(cast):
   return ast.literal_eval(cast)
```

```python
# Store filepath in a variable
credits_file = "Movies/credits.csv"
```

```python
credits_df = pd.read_csv(credits_file)
credits_df.head()
```

```python
# creates a version of the cast column that allows retrieval by character tag
credits_df["cast_new"] = credits_df["cast"].apply(transform_cast)
credits_df.head()
```

```python
# create character and movie id lists by looping through each row and through all characters within each cast_new 
character_list = []
movie_list = []
for i in range(len(credits_df)):
    movie_id = credits_df.iloc[i,2]
    for x in credits_df.iloc[i,3]:
        character_list.append(x['character'])
        movie_list.append(movie_id)
```

```python
# combine both lists
character_movie = zip(character_list, movie_list)
```

```python
# dataframe with movie character and movie id
labels = ['character', 'id']
movie_character_df = pd.DataFrame.from_records(character_movie, columns=labels)
movie_character_df.head()
```

```python
movie_character_df[['MOVIE_CHARACTER','ALIAS']] = movie_character_df.character.str.split("\s+\(", expand=True, n=1)
movie_character_df.head()
```

```python
movie_character_df = movie_character_df.drop(['ALIAS'], axis=1)
movie_character_df.head()
```

```python
movie_character_df.to_csv('output/movie_character.csv', index = None, header=True)
```
